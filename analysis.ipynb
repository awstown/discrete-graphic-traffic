```python
import traffic_objects as toimport rules as rimport numpy as npimport matplotlib.pyplot as plt
```

```python
## Create constantsvmax = 3carRange = range(100)
```

```python
## Create Data lists for each modelSTCAdata  = []ASEPdata  = []CA184data = []
```

```python
# STCA Datafor i in carRange:    a = to.Data()    lane = to.Lane(100)    lane.populate(i+1)    r.stca(a, lane, vmax)    STCAdata.append(a)# ASEP Datafor i in carRange:    a = to.Data()    lane = to.Lane(100)    lane.populate(i+1)    r.asep(a, lane, vmax)    ASEPdata.append(a)# CA-184 Datafor i in carRange:    a = to.Data()    lane = to.Lane(100)    lane.populate(i+1)    r.ca184(a, lane, vmax)    CA184data.append(a)        
```

```python
def process(data):    current = []    numOfCars = []    for run in data:        meanOfAll = []        for car_speed in run.speed_history:            meanOfAll.append(np.mean(car_speed))        current.append(mean(meanOfAll))        numOfCars.append(run.number_of_cars)    return (numOfCars, current)STCAplotData  = process(STCAdata)ASEPplotData  = process(ASEPdata)CA184plotData = process(CA184data)
```

```python
plt.plot(STCAplotData[0],STCAplotData[1])
```
